```python
import requests
import json
```

```python
# URL untuk mendapatkan daftar postingan
api_url = "https://jsonplaceholder.typicode.com/posts"

try:
    # Lakukan permintaan GET untuk mendapatkan daftar postingan
    response = requests.get(api_url)

    # Memeriksa apakah permintaan berhasil
    response.raise_for_status()  # Memicu exception jika status code bukan 200

    # Mengonversi respons ke JSON
    posts = response.json()  # Mengakses hasil JSON

    # Tampilkan daftar postingan (dataset)
    print("Daftar Postingan:")
    for idx, post in enumerate(posts):
        print(f"{idx + 1}. {post['title']} (ID: {post['id']})")
    
except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")  # Tampilkan kesalahan HTTP
except requests.exceptions.RequestException as req_err:
    print(f"Request error occurred: {req_err}")  # Tampilkan kesalahan permintaan
except ValueError as json_err:
    print(f"JSON decode error: {json_err}")  # Tampilkan kesalahan decode JSON
```

```python
# Meminta pengguna untuk memasukkan ID postingan yang diinginkan
post_id = int(input("Masukkan ID postingan yang ingin diambil (1-100): "))

# Memeriksa apakah ID valid
if 1 <= post_id <= 100:
    selected_post = next((post for post in posts if post['id'] == post_id), None)
    if selected_post:
        print(f"Post yang dipilih: {selected_post['title']} (ID: {selected_post['id']})")
    else:
        print("Post tidak ditemukan.")
else:
    print("ID tidak valid. Harap masukkan ID antara 1 dan 100.")
```

```python
# Menampilkan detail dari postingan yang dipilih
if selected_post:
    print("\nDetail Postingan:")
    print(f"ID: {selected_post['id']}")
    print(f"Judul: {selected_post['title']}")
    print(f"Isi: {selected_post['body']}")
```

```python
# Menyimpan detail postingan ke dalam file JSON
if selected_post:
    with open(f"post_{post_id}.json", 'w') as json_file:
        json.dump(selected_post, json_file, indent=4)
    print(f"Detail postingan telah disimpan ke 'post_{post_id}.json'.")
```
